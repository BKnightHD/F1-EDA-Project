```python
# Functions

class operations:

  @staticmethod
  def add(x, y):
    result = int(x) + int(y)
    return result

  @staticmethod
  def subtract(x, y):
    result = int(x) - int(y)
    return result

  @staticmethod
  def multiply(x, y):
    result = int(x) * int(y)
    return result

  @staticmethod
  def divide(x, y):
      try:
          result = int(x) / int(y)
          print(f"{x} / {y} = {result}")
          return result
      except ZeroDivisionError:
          print(f"{x} / {y} is undefined")
          return None

# Prmary Code

while True:
  # Get first number
  first_number = input("Enter a number: ")
  while True:
    try:
      new_number = float(first_number)
      break # exit loop when succesful
    except ValueError:
      first_number = input("This is not a number. Please enter a number or type 'q' to quit: ")
      if first_number.lower() == 'q':
        break

  if first_number.lower() == 'q': # stops program if q is selected
    break

  # Get operator
  operator_list = ['+', '-', '*', '/']
  operator = input("Enter an operator (+, -, *, /): ")
  while True:
    if operator not in operator_list:
      operator = input("Input invalid, please enter a correct operator (+, -, *, /) or type 'q' to quit: ")
      if operator.lower() == 'q': # stops program if q is selected
        break

    else:
      break

  if operator.lower() == 'q':
    break

  # Get second number
  second_number = input("Enter another number: ")
  while True:
    try:
      new_number = float(second_number)
      break # exit loop when succesful
    except ValueError:
      second_number = input("This is not a number. Please enter a number or type 'q' to quit: ")
      if second_number.lower() == 'q': # stops program if q is selected
        break

  if second_number.lower() == 'q': # stop program is q is selected
    break


# Perform Calculations

  if operator == "+":
    new_result = operations.add(first_number, second_number)
    print(f"{first_number} {operator} {second_number} = {new_result}")

  elif operator == "-":
    new_result = operations.subtract(first_number, second_number)
    print(f"{first_number} {operator} {second_number} = {new_result}")

  elif operator == "*":
    new_result = operations.multiply(first_number, second_number)
    print(f"{first_number} {operator} {second_number} = {new_result}")

  elif operator == "/":
    operations.divide(first_number, second_number)

  # Ask if user wants to continue
  continue_choice = input("\nDo you want to perform another calculation? (y/n): ")
  if continue_choice.lower() != 'y':
    break

print("Calculator exited. Goodbye!")

```
