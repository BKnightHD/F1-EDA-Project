```python
import requests

api_url = 'https://fakerapi.it/api/v2/creditCards?_quantity=500' # Define the API endpoint

response = requests.get(api_url) # Make the GET request
```

```python
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Access the response data, typically in JSON format
    data = response.json()
    print("Data retrieved successfully:")

else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
df = pd.DataFrame(data['data'])
```
